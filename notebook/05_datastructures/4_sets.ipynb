```python
# 2つの単語の一意の文字に対する集合演算のデモ
a = set('abracadabra')
b = set('alacazam')
a                                  # a の一意の文字
```

```python
a - b                              # a にはあるが b にはない文字
```

```python
a | b                              # a または b または両方の文字
```

```python
a & b                              # a と b の両方に存在する文字
```

```python
a ^ b                              # a または b にはあるが両方にはない文字
```

```python
a = {x for x in 'abracadabra' if x not in 'abc'}
a
```
