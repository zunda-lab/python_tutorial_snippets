```python
word = 'Python'
word[0]  # 位置 0 の文字
```

```python
word[5]  # 位置 5 の文
```

```python
word[-1]  # 最後の文字
```

```python
word[-2]  # 最後から2番目の文字
```

```python
word[-6]
```

```python
word[0:2]  # 位置 0 (含む) から 2 (含まない) までの文字
```

```python
word[2:5]  # 位置 2 (含む) から 5 (含まない) までの文字
```

```python
word[:2]  # 先頭から位置 2 (含まない) までの文字
```

```python
word[4:]  # 位置 4 (含む) から末尾までの文字
```

```python
word[-2:]  # 最後から2番目 (含む) から末尾までの文字
```

```python
word[:2] + word[2:]
```

```python
word[:4] + word[4:]
```
