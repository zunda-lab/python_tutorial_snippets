```python
try:
    raise KeyboardInterrupt
finally:
    print('Goodbye, world!')
```

```python
def bool_return():
    try:
        return True
    finally:
        return False

bool_return()
```

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("division by zero!")
    else:
        print("result is", result)
    finally:
        print("executing finally clause")

divide(2, 1)
```

```python
divide(2, 0)
```

```python
divide("2", "1")
```
