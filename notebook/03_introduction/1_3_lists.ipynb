```python
cubes = [1, 8, 27, 65, 125] # ここに何か間違いがあります
4 ** 3  # 4の3乗は64であって、65ではありません!
```

```python
cubes[3] = 64  # 間違った値を置き換えます
cubes
```

```python
cubes.append(216)  # 6の3乗を追加します
cubes.append(7 ** 3)  # そして7の3乗を追加します
cubes
```

```python
rgb = ["Red", "Green", "Blue"]
rgba = rgb
id(rgb) == id(rgba)  # それらは同じオブジェクトを参照します
```

```python
rgba.append("Alph")
rgb
```

```python
correct_rgba = rgba[:]
correct_rgba[-1] = "Alpha"
correct_rgba
```

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
letters
```

```python
# いくつかの値を置き換えます
letters[2:5] = ['C', 'D', 'E']
letters
```

```python
# 今度はそれらを削除します
letters[2:5] = []
letters
```

```python
# すべての要素を空のリストに置き換えることでリストをクリアします
letters[:] = []
letters
```
