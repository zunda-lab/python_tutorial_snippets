```python
import os
import glob
from pprint import pprint
import json
import re
```

```python
os.chdir("/content/drive/MyDrive/Colab Notebooks/python_activities/python_tutorial_snippets/notebook")
r_fnames = glob.glob('*.ipynb')
print(r_fnames)
```

```python
for r_fname in r_fnames:
    # print(fname)
    result = ""
    basename_without_ext = os.path.splitext(os.path.basename(r_fname))[0]
    w_fname = basename_without_ext + ".py"
    with open(r_fname, "r") as f:
        nb = json.load(f)
        # pprint(nb)
        for cell in nb["cells"]:
            for source_line in cell["source"]:
                result += source_line
            result += "\n"
            for output in cell["outputs"]:
                output_type = output["output_type"]
                match output_type:
                    case "execute_result":
                        for text_line in output["data"]["text/plain"]:
                            result += f"# {text_line}\n"
                    case "display_data":
                        for text_line in output["data"]["text/plain"]:
                            result += f"# {text_line}\n"
                    case "stream":
                        for text_line in output["text"]:
                            result += f"# {text_line}"
                    case "error":
                        for text_line in output["traceback"]:
                            text_line = re.compile(r'\x1B(?:[@-Z\\-_]|\[[0-?]*[ -/]*[@-~])').sub('', text_line)
                            if not text_line.endswith("\n"):
                                text_line += "\n"
                            result += f"# {text_line}"
                result += "\n"
    print(result)
    with open(w_fname, "w") as f:
        f.write(result)
```
