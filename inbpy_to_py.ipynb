```python
import os
import glob
from pprint import pprint
import json
import re
```

```python
os.chdir("/content/drive/MyDrive/Colab Notebooks/python_activities/python_tutorial_snippets/notebook/09_classes")
read_fpaths = glob.glob('./**/*.ipynb', recursive=True)
pprint(read_fpaths)
```

```python
for read_fpath in read_fpaths:
    # print(fname)
    result = ""
    dir_path = os.path.dirname(read_fpath)
    basename_without_ext = os.path.splitext(os.path.basename(read_fpath))[0]
    write_fpath = f"{dir_path}/{basename_without_ext}.py"
    print(write_fpath)
    with open(read_fpath, "r") as f:
        nb = json.load(f)
        # pprint(nb)
        for cell in nb["cells"]:
            for source_line in cell["source"]:
                result += source_line
            result += "\n"
            for output in cell["outputs"]:
                output_type = output["output_type"]
                match output_type:
                    case "execute_result":
                        for text_line in output["data"]["text/plain"]:
                            result += f"# {text_line}\n"
                    case "display_data":
                        for text_line in output["data"]["text/plain"]:
                            result += f"# {text_line}\n"
                    case "stream":
                        for text_line in output["text"]:
                            result += f"# {text_line}"
                    case "error":
                        for text_line in output["traceback"]:
                            text_line = re.compile(r'\x1B(?:[@-Z\\-_]|\[[0-?]*[ -/]*[@-~])').sub('', text_line)
                            if not text_line.endswith("\n"):
                                text_line += "\n"
                            result += f"# {text_line}"
            result += "\n"
    # print(result)
    with open(write_fpath, "w") as f:
        f.write(result)
```
